1. What is a Vector?
```
Vector: a measure or quantity that has magnitude & direction (+ or -).
```

2. What does the norm of a vector represent?
```
The length or magnitude of the vector, by finding the sum 
of each component squared and then taking the square root of that sum
```

3. What makes a Vector Orthogonal?
```
The 2 vectors are orthogonal (90 degree) to each other if the dot product is zero. They are linearly independent
```

4. What is Principal component analysis or PCA?
```
PCA is a method to interpret data by reducing dimensionality of a large data table withoiut losing the important information. 
It's like a"summary" of a big table that can be easily visualized and analysed.
PCA reduces data by geometrically projecting them onto lower dimensions called principal components (PCs), 
with the goal of finding the best summary of the data using a limited number of PCs
```

5. What is K-Means?
```
K-means is one of clustering methods. The goal is to get similar data points together into a fix number (=k) of groups and discover underlying pattern. 
```

6. What are Eigenvalues?
```
Eigenvalue is a set of numbers in linear system that tell us how much variance is in the data,  
or how spread out is the data on certain direction.
```

7. What are Eigenvectors?
```
Eigenvector is a non-zero vector that says there is a scalar associated with that vector
```

# **Graph vector**

````
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
````

\begin{align}
\vec{a} = \begin{bmatrix} 9 \\ 6 \end{bmatrix}
\end{align}

\begin{align}
\vec{b} = \begin{bmatrix} 12 \\ 9 \end{bmatrix}
\end{align}

\begin{align}
\vec{c} = \begin{bmatrix} -4 \\ 3 \end{bmatrix}
\end{align}

````
a = [9, 6]
b = [12, 9]
c = [-4, 3]

plt.arrow(0,0, a[0], a[1],head_width=.1, head_length=0.1, color ='blue')
plt.arrow(0,0, b[0], b[1],head_width=.1, head_length=0.1, color ='red')
plt.arrow(0,0, c[0], c[1],head_width=.1, head_length=0.1, color ='black')

plt.style.use('ggplot')

plt.xlim(-5,13)          
plt.ylim(-5,10)
plt.title("a (blue), b (red), c (black) ")
plt.legend();
plt.show()
````

\begin{align}
{d} = 
\begin{bmatrix}
    9 & 12 & 6 \\
    6 & 3 & 9 \\
    1 & 4 & 8
\end{bmatrix}
\end{align}

````
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vectors = np.array([[0, 0, 0, 9, 6, 1], 
                    [0, 0, 0, 12, 3, 4],
                    [0, 0, 0, 6, 9, 8]])

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 13])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
````
